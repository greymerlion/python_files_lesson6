## Assessment 3 - Number Guesser

This program should play a number guessing game.

Here are the rules:

1) The computer picks a number between 1 and 100, and doesn't tell user

2) The player is given a chance to guess the number via input

   * If the player guesses correctly the game is over and the count of how many guesses they took is printed out. Ask the player if they would like to play again and if so, return to step 1 above, otherwise stop the game.

   * If the player guesses incorrectly the computer will tell the user whether their number is too high or too low. and the player is allowed to guess again, return to step 2 above
